```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'axes.titlesize': 24,
             'axes.labelsize': 20,
             'xtick.labelsize': 16,
             'ytick.labelsize': 16})
```

```python
x = np.linspace(-2.5, 2.5, 100)
y = x**2

x_0 = 2
y_0 = x_0**2

x_1 = 1.6
y_1 = x_1**2

x_2 = 1.28
y_2 = x_2**2

sns.set_style('whitegrid')

plt.plot(x, y)

plt.scatter(x_0, y_0, label = r'$x_0$', s = 100, color = 'black')
plt.scatter(x_1, y_1, label = r'$x_1$', s = 100, color = 'black', marker = ',')
plt.scatter(x_2, y_2, label = r'$x_2$', s = 100, color = 'black', marker = 'v')

plt.title(r'Gradient Descent for $\eta = 0.1$')
plt.legend(fontsize = 14, loc = 'lower left')
plt.tight_layout()

plt.savefig('Gradient Descent', dpi = 300)
```
