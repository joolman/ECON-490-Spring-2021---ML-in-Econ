# Splitting GapMinder

```python
import pandas as pd
```

```python
gm = pd.read_csv('gapminder.csv')
```

```python
gm.head()
```

```python
gm.columns
```

```python
gm_gdp = gm.loc[:, gm.columns.str.startswith('co') | gm.columns.str.startswith('gdp')]
```

```python
gm_life = gm.loc[:, gm.columns.str.startswith('co') | gm.columns.str.startswith('life')]
```

```python
gm_pop = gm.loc[:, gm.columns.str.startswith('co') | gm.columns.str.startswith('pop')]
```

## Adjusting data 

```python
# Leave GDPc alone
gm_gdp.head()
```

```python
# Leave life Expecatancy Alone
gm_life.tail()
```

```python
gm_pop.head()
```

```python
gm_pop = pd.wide_to_long(gm_pop, stubnames = 'pop', i = ['country', 'continent'], j = 'year', sep = '_').reset_index()
```

```python
gm_pop.columns = gm_pop.columns.str.upper()
```

```python
gm_pop.head()
```

## Saving 

```python
gm_gdp.to_csv('gapminder GDPc.csv', index = False)
```

```python
gm_life.to_csv('gapminder life expectancy.csv', index = False)
```

```python
gm_pop.to_csv('gapminder population.csv', index = False)
```
