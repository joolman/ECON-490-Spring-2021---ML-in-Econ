```python
# utilities
import numpy as np

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'axes.titlesize': 24,
             'axes.labelsize': 20,
             'xtick.labelsize': 12,
             'ytick.labelsize': 12,
             'figure.figsize': (8, 4.5),
             'figure.dpi': 160})
sns.set_style('white')
```

```python
lims = 9
x = np.linspace(-lims, lims, num = 100)
gamma1 = 0.1
y1 = np.exp(-gamma1*(x)**2)

gamma2 = 5
y2 = np.exp(-gamma2*(x)**2)


plt.plot(x, y2, linewidth = 4, label = '$\\gamma$ = {0:d}'.format(gamma2))
plt.plot(x, y1, linewidth = 4, label = '$\\gamma$ = {0:.1f}'.format(gamma1))

plt.legend()
plt.yticks([])
plt.xticks([])

sns.despine()
plt.savefig('gamma', dpi = 160)
```
