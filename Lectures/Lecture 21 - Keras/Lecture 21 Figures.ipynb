```python
import numpy as np

from tensorflow.keras.activations import sigmoid, tanh, softplus, relu, elu

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (16/3, 9/3))

plt.subplot(2, 2, 1)
x = np.linspace(-4, 4, 100)
y = sigmoid(x)
plt.plot(x, y)
plt.grid()
plt.ylim(-2, 2)
plt.yticks([-2, -1, 0, 1, 2])
plt.title('"sigmoid"')

plt.subplot(2, 2, 2)
y = tanh(x)
plt.plot(x, y)
plt.grid()
plt.ylim(-2, 2)
plt.yticks([-2, -1, 0, 1, 2])
plt.title('"tanh"')

plt.subplot(2, 2, 3)
y = softplus(x)
plt.plot(x, y)
plt.grid()
plt.ylim(-1, 3)
plt.yticks([-1, 0, 1, 2, 3])
plt.title('"softplus"')

plt.subplot(2, 2, 4)
y = relu(x)
plt.plot(x, y)
plt.grid()
plt.ylim(-1, 3)
plt.yticks([-1, 0, 1, 2, 3])
plt.title('"relu"')

plt.tight_layout()

plt.savefig('activation functions 1', dpi = 300)
```

```python
plt.figure(figsize = (16/3, 9/3))

plt.subplot(2, 1, 1)
y = relu(x, alpha = 0.1)
plt.plot(x, y)
plt.grid()
plt.ylim(-1, 3)
plt.yticks([-1, 0, 1, 2, 3])
plt.title('"relu(x, alpha = 0.1)"')

plt.subplot(2,1, 2)
y = elu(x, alpha = 1)
plt.plot(x, y)
plt.grid()
plt.ylim(-1, 3)
plt.yticks([-1, 0, 1, 2, 3])
plt.title('"elu(x, alpha = 1)"')

plt.tight_layout()
plt.savefig('activation functions 2', dpi = 300)
```

```python
x = np.linspace(0, 100, num = 100)
eta_0 = 1
decay_steps = 10

y_p = eta_0/(1+x/decay_steps)

decay_rate = 0.1
y_e1 = eta_0*decay_rate**(x/decay_steps)
decay_rate = 0.9
y_e2 = eta_0*decay_rate**(x/decay_steps)

plt.figure(figsize = (9/3, 9/3))

plt.plot(x, y_p, label = 'Power Decay')
plt.plot(x, y_e1, linestyle = 'dashed', label = 'Exp. r = 0.1')
plt.plot(x, y_e2, linestyle = 'dotted', label = 'Exp. r = 0.9')

plt.title('s = 10')
plt.yticks((1, 0), (r'$\eta_0$', 0))
plt.legend()
plt.tight_layout()

plt.savefig('learning schedules', dpi = 300)
```
