```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
gdp = pd.read_csv('GDPC1.csv', index_col = 'DATE')
urate = pd.read_csv('UNRATE.csv', index_col = 'DATE')
sp = pd.read_csv('sp500.csv', index_col = 'Date', usecols = ['Date', 'Close'])
sp = sp['Close']
sp.sort_index(inplace = True)
```

```python
ss = pd.read_csv('SN_d_tot_V2.0.csv', 
           sep = ';',
           header = None,
           names = ['year', 'month', 'day', 'num_date', 'count', 'sd', 'drop1', 'drop2'])
ss['Date'] = pd.to_datetime(ss[['year', 'month', 'day']])
ss.set_index('Date', inplace = True)
ss = ss['count']
ss = ss[ss != -1]
```

```python
temp = pd.read_csv('jena_climate_2009_2016.csv', 
           index_col = 'Date Time',
           usecols = ['Date Time', 'T (degC)'])
```

```python
ax = gdp.plot(legend = False,
        figsize = (16/3, 9/3))

plt.grid()
plt.xticks(rotation = 25)
plt.ylabel('GDP (2012 $bn)')
plt.xlabel('')
plt.title('Quarterly rGDPc n: {}'.format(gdp.shape[0]))
plt.tight_layout()

fig = ax.get_figure()
fig.savefig('gdp', dpi = 300)
```

```python
ax = urate.plot(legend = False,
        figsize = (16/3, 9/3))

plt.grid()
plt.xticks(rotation = 25)
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('')
plt.title('Monthly U-Rate n: {}'.format(urate.shape[0]))
plt.tight_layout()

fig = ax.get_figure()
fig.savefig('urate', dpi = 300)
```

```python
ax = sp.plot(legend = False,
        figsize = (16/3, 9/3))

plt.grid()
plt.xticks(rotation = 25)
plt.ylabel('Index Value')
plt.xlabel('')
plt.title('Daily without weekends or holidays n: {}'.format(sp.shape[0]))
plt.tight_layout()

fig = ax.get_figure()
fig.savefig('snp', dpi = 300)
```

```python
ax = ss.plot(legend = False,
        figsize = (16/3, 9/3))

plt.grid()
plt.xticks(rotation = 25)
plt.ylabel('Count')
plt.xlabel('')
plt.title('~Daily Sun Spots n: {}'.format(ss.shape[0]))
plt.tight_layout()


fig = ax.get_figure()
fig.savefig('sun spots', dpi = 300)
```

```python
ax = temp.plot(legend = False,
        figsize = (16/3, 9/3))

plt.grid()
plt.xticks(rotation = 25)
plt.ylabel('Temperature (C)')
plt.xlabel('')
plt.title('Every 10 Minute Temperature n: {}'.format(temp.shape[0]))
plt.tight_layout()

fig = ax.get_figure()
fig.savefig('Temperature', dpi = 300)
```
